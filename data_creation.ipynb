```python
x1=np.empty((200,64,64),dtype=float)
y1=np.empty(200,dtype=int)
x2=np.empty((200,64,64),dtype=float)
y2=np.empty(200,dtype=int)
```

```python
j=k=0
for i in range(400):
    if target[i]<20:
        x1[j]=data[i]
        y1[j]=target[i]
        j+=1
    else:
        x2[k]=data[i]
        y2[k]=target[i]
        k+=1
        
del data,target
```

```python
print(x1.shape)
from PIL import Image
img=Image.fromarray(np.reshape(x1[0]*255,(64,64)))
img.show()
k
```

```python
x_train=np.empty((20100,2,64,64),dtype=float)
y_train=np.empty((20100,),dtype=int)
```

```python
k=0
for i in range(200):
    for j in range(i,200):
        x_train[k][0]=x1[i]
        x_train[k][1]=x1[j]
        if y1[i]==y1[j]:
            y_train[k]=1
        else:
            y_train[k]=0
        k+=1
        print(i,j,k)
del x1,y1
    
```

```python
x_test=np.empty((20100,2,64,64),dtype=float)
y_test=np.empty((20100,),dtype=int)

```
